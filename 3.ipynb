```python
#23A91AO4E8
#P.HARSHIT

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# -------------------------------------------------------
# 1) LOAD IRIS CSV FILE
# -------------------------------------------------------
df = pd.read_csv("/content/Iris.csv")
print(df.head())
# X contains input features
X = df.drop("Species", axis=1)
# y contains output class
y = df["Species"]
```

```python
# -------------------------------------------------------
# 2) STANDARDIZE THE FEATURES
# -------------------------------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled[:5])
```

```python
# -------------------------------------------------------
# 3) APPLY PCA (reduce to 2 components)
# -------------------------------------------------------
pca2 = PCA(n_components=2)
X_pca2 = pca2.fit_transform(X_scaled)
print("\nExplained Variance Ratio (2 Components):")
print(pca2.explained_variance_ratio_)
```

```python
print("Total variance explained:", pca2.explained_variance_ratio_.sum())
```

```python
print(X_pca2[:5])  # first 5 rows
```

```python
# -------------------------------------------------------
# 4) 2D PCA SCATTER PLOT
# -------------------------------------------------------
plt.figure(figsize=(8,6))
for species in np.unique(y):
    plt.scatter(
        X_pca2[y == species, 0],
        X_pca2[y == species, 1],
        label=species,
        s=60
)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Iris Dataset (2D)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```


For a 3D PCA scatter plot, you need to use a 3D projection. Here's the corrected code:


```python
# -------------------------------------------------------
# 3) APPLY PCA (reduce to 3 components)
# -------------------------------------------------------
pca3 = PCA(n_components=3)
X_pca3 = pca3.fit_transform(X_scaled)
print("\nExplained Variance Ratio (3 Components):")
print(pca3.explained_variance_ratio_)
```

```python
print("Total variance explained:", pca3.explained_variance_ratio_.sum())
```

```python
print(X_pca3[:10])  # first 10 rows
```

```python
# -------------------------------------------------------
# 3D PCA SCATTER PLOT
# -------------------------------------------------------
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting

# Create a 3D figure
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')  # Create 3D axes

for species in np.unique(y):
    ax.scatter(
        X_pca3[y == species, 0],  # x-coordinate (PC1)
        X_pca3[y == species, 1],  # y-coordinate (PC2)
        X_pca3[y == species, 2],  # z-coordinate (PC3)
        label=species,
        s=60
)

ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("PCA on Iris Dataset (3D)")
plt.legend()
plt.tight_layout()
plt.show()
```
